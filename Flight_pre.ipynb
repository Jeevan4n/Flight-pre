```python
ab['Date']=ab['Date_of_Journey'].str.split('/').str[0]
ab['Month']=ab['Date_of_Journey'].str.split('/').str[1]
ab['Year']=ab['Date_of_Journey'].str.split('/').str[2]
ab.drop('Date_of_Journey',axis=1,inplace=True)
ab.head(5)
```

```python
ab['Date']=ab['Date'].astype(int)
ab['Month']=ab['Month'].astype(int)
ab['Year']=ab['Year'].astype(int)
ab.info()
```

```python
ab['Arrival_Time']=ab['Arrival_Time'].str.split(' ').str[0]
#by using lambda functions
# ab['Arrival_Time].apply(lamda x:x.split(' ')[0])
```

```python
ab['Arrival_hour']=ab['Arrival_Time'].str.split(':').str[0]
ab['Arrival_min']=ab['Arrival_Time'].str.split(':').str[1]
ab.drop('Arrival_Time',axis=1,inplace=True)
ab.head(5)
```

```python
ab['Arrival_hour']=ab['Arrival_hour'].astype(int)
ab['Arrival_min']=ab['Arrival_min'].astype(int)
ab.info()
```

```python
ab['Dept_hour']=ab['Dep_Time'].str.split(':').str[0]
ab['Dept_min']=ab['Dep_Time'].str.split(':').str[1]
ab.drop('Dep_Time',axis=1,inplace=True)
ab['Dept_hour']=ab['Dept_hour'].astype(int)
ab['Dept_min']=ab['Dept_min'].astype(int)
ab.info()
```

```python
ab['Total_Stops']=ab['Total_Stops'].map({'non-stop':0,'1 stop':1,'2 stops':2,'3 stops':3,'4 stops':4})
ab['Total_Stops']=ab['Total_Stops'].fillna(1)
ab['Total_Stops']=ab['Total_Stops'].astype(int)
ab.info()
```

```python
ab.drop('Route',axis=1,inplace=True)
ab.head()
```

```python
ab['Additional_Info'].unique()
```

```python
ab['Duration'] = ab['Duration'].astype(str)
# boolean indexing to filter out rows where 'Duration' is '5m'
ab = ab[ab['Duration'] != '5m']
# modifications that would have been occured in the duration column please do analysis
ab['Duration_hour'] = ab['Duration'].str.split(' ').str[0].str.split('h').str[0]
ab['Duration_hour'] = ab['Duration_hour'].astype(int)
ab['Duration_hour']=ab['Duration_hour']*60
ab['Duration_min'] = ab['Duration'].str.split(' ').str[1].str.split('m').str[0]
ab['Duration_min'] = ab['Duration_min'].fillna(0)
ab['Duration_min']=ab['Duration_min'].astype(int)
ab['Duration_in_min']=ab['Duration_hour']+ab['Duration_min']
ab.drop('Duration',axis=1,inplace=True)
ab.drop('Duration_hour',axis=1,inplace=True)
ab.drop('Duration_min',axis=1,inplace=True)
ab.head()
```

```python
ab['Airline'].unique()
```

```python
from sklearn.preprocessing import LabelEncoder
Le=LabelEncoder()
```

```python
ab['Airline']=Le.fit_transform(ab['Airline'])
ab['Source']=Le.fit_transform(ab['Source'])
ab['Destination']=Le.fit_transform(ab['Destination'])
ab['Additional_Info']=Le.fit_transform(ab['Additional_Info'])
```

```python
# Import Libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error

# Load Data (replace 'your_dataset.csv' with your file path)
data = pd.read_csv('Flight_price.csv')

# Split the Data into Rows with Known and Unknown Prices
known_price_data = data[data['Price'].notnull()]  # Rows where Price is known
unknown_price_data = data[data['Price'].isnull()]  # Rows where Price is missing

# Separate Features and Target in the Known Price Data
X_train = known_price_data.drop('Price', axis=1)
y_train = known_price_data['Price']

# Use the Known Data to Train the Model
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# Prepare the Features for the Unknown Price Data
X_unknown = unknown_price_data.drop('Price', axis=1)

# Predict Missing Prices
predicted_prices = model.predict(X_unknown)

# Fill in the Predicted Prices in the Original DataFrame
data.loc[data['Price'].isnull(), 'Price'] = predicted_prices

# (Optional) Print the updated DataFrame with Predicted Prices filled in
print(data.head())
```
